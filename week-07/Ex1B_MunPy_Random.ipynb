## Genrating Random Choices from an Array

```python
students = np.array(['Luis', 'Charles', 'Corey', 'Blake', 'Alexus', 'Alondra', 'Daniel', 'Victoria', 'Hana', 'Lorah', 'Sonia', 'Samia'])
```

```python
call_on = np.random.choice(students)
print(call_on)
```

```python
breakout_groups = np.random.choice(students, size =(3, 4))
print(breakout_groups)
# This would NOT be effective for breakout groups because the same student can appear more than once in the results
```

```python
call_on_weighted = np.random.choice(students, size=5, p=[
    .30,
    .10,
    .10,
    .10,
    .10,
    .05,
    .05,
    .05,
    .05,
    .05,
    .05,
    .00
])
print(call_on_weighted)
#p values must add up to exactly 1.0
# Samia has 0% chance of being selected
# Luis has the highest chance at 30%
```

## Generating a Normal Distribution

```python
bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))
```

```python
print(bell_curve)
# loc = mean(peak of the curve) = 10
# scale = standard deviatoin = 5
# size = 4 rows, 5 columns
```
