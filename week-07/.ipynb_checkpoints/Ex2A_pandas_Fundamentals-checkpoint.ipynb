# Exercise 2.A Fundamentals of pandas

```python
import numpy as np
import pandas as pd
```

## Series

```python
aa = [1, 2, 3, 4]
ab = (5, 6, 7, 8)
ac = list({9, 10, 11, 12})
ad = {'first': 13, 'second': 14, 'third': 15, 'fourth': 16}
```

```python
ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
ser3 = pd.Series(ac)
ser4 = pd.Series(ad)

print(ser1)
print(ser2)
print(ser3)
print(ser4)
```

```python

try:
    ser5 = pd.Series(17, 18, 19, 20)
except TypeError as e:
    print(f"TypeError: {e}")

#TypeError: got an unexpected keyword argument
#pd.Series doesnt accept multiple positional argumetns like this
```

## DataFrames from Series

```python
df1 = pd.DataFrame(ser1)
df1
```

```python
df1.insert(1, 'col2', ser2)
df1.insert(2, 'col3', ser3)
df1
```

```python
df1['col4'] = ser4
df1
#ser4 uses string keys
#instead of numbers
# pandas cant align values and fills with NaN
```

```python
df1.loc[4] = [5, 9, 12, 0]
df1
```

## Coding a DataFrame

```python
test_data = {
    'ID' : [1, 2, 3, 4, 5, 6, 7, 8],
    'Name' : ['Alicia', 'Bob', 'Charlie', 'Dominic', 'Eve', 'Frank', 'Grace', 'Heidi'],
    'Age' : [25, np.nan, 22, 19, 31, 35, np.nan, 18],
    'City': ['NY', 'LA', 'Chicago', 'Houston', 'Phoenix', 'Boston', 'Austin', 'San Diego'],
    'Score': [85, 92, -1, 65, 78, np.nan, 55, 90]
}
```
