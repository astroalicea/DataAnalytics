# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

## Looping Through Arrays

```python
arrA = np.array([(11, 12, 13), (14, 15, 16)])
```

```python
print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

The first loop multiplies each row by 2, returning arrays.
The second loop multiplies each individual element by 2 usnig a nested loop, returning single numbers.

```python
for x in arrA:
    for y in x:
        loop_result = x * 2
        print(loop_result)
#loop_result miltiplies the whole row x by 2, not just y so even though we loop through each element with y
#prints entire row 3 times.
```

## Join and Stacking Arrays

```python
joined_array_1 = np.concatenate((arrA, arrA), axis=0)
print(joined_array_1)
```

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array_1 =np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)
```

```python
stacked_array_2 = np.stack((joined_array_1, joined_array_2))
print(stacked_array_2)

#ValueError: all input arrays must have the same shape. joined_array_1 has 4 rows but arrA has 2 rows
```

```python
h = np.hstack((arrA, arrA))
print(h)
#hstack horizontally - same as axis=1, adds columns

v = np.vstack((arrA, arrA))
print(v)
#vstack stacks vertically - same as axis=0, adds rows

d = np.dstack((arrA, arrA))
print(d)
#dstack stacks depth-wise - adds a third dimension
```

## Splitting Arrays

```python
splitsky = np.array_split(joined_array_1, 2)
print(splitsky)
#Result: 2 arrays, each with 2 rows and 3 columns(2D)
```

```python
spliterino = np.array_split(joined_array_1, 3)
print(spliterino)
#Result: 3 arrays - first two have 2 rows, last has 0 rows (2D each)
```

## Searching and Replacing Array Values

```python
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')
print(result_1)
# % 2 == 0 checks if number is even
# if even: keep the vlaue from arrA
#if odd: replace with 'odd'
```

## Sorting and Filtering Arrays

```python
arrB = np.array([(7, 6), (13, 5), (7, 3)])
arrB_sorted = np.sort(arrB)
print(arrB_sorted)
#sorts each row from smallest to largest
```

```python
print(arrB)
z = arrB < 7
arrC = arrB[z]
print(arrC)
#z is a boolean array - True where value < 7, False otherwise
#arrC contains only values where z is True
```

```python
y = arrB >= 7
arrD = arrB[y]
print(arrD)
# y is True where value >= 7
# arrD contains only those values
```
