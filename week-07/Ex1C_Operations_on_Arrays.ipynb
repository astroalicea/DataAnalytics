# Exercise 1.C Performing operations on arrays

```python
import numpy as np
```

## Sample Arrays

```python
r_1 = np.random.randint(1, 20, size=2)
r_2 = np.random.randint(1, 20, size=(1, 2))
r_3 = np.random.rand(2, 1)
r_4 = np.random.randint(20, 100, size=(4, 5))
r_5 = np.random.rand(4, 5)
```

## Permutations

```python
print(r_4)
```

```python
np.random.shuffle(r_4)
print(r_4)
```

The shuffle function rearranged the rows of r_4 but kept the values within each row in the same order. It modifies the original
array directly.

```python
new_shuffle = np.random.permutation(r_4)
print(new_shuffle)
print(r_4)
```

Use shuffle when you want to modify the original array in place.
Use permutation when you want a shuffled copy but need the original.

## Mathematical Operations

```python
print(f"r_1: {r_1}, shape: {r_1.shape}")
print(f"r_2: {r_2}, shape: {r_2.shape}")
result = np.add(r_1, r_2)
print(f"result: {result}, shape: {result.shape}")
```

```python
print(f"r_1: {r_1}, shape: {r_1.shape}")
print(f"r_3: {r_3}, shape: {r_3.shape}")
result2 = np.add(r_1, r_3)
print(f"result: {result2}, shape: {result2.shape}")
```

```python
print(f"r_2: {r_2}, shape: {r_2.shape}")
print(f"r_4: {r_4}, shape: {r_4.shape}")
result3 = np.add(r_2, r_4)
print(f"result: {result3}, shape: {result3.shape}")
```

```python
try:
    print(np.add(r_1, r_4))
except ValueError as e:
    print(f"r_1 + r_4 error: {e}")
# r_1 is hsape (2,) and r_4 is (4,5) - cannot broadcast together

try:
    print(np.add(r_2, r_4))
except ValueError as e:
    print(f"r_2 + r_4 error: {e}")

try:
    print(np.add(r_3, r_4))
except ValueError as e:
    print(f"r_3 + r_4 error: {e}")
```

```python
print(np.add(np.array([1]), r_4))

print(np.add(np.array([[1]]), r_4))

print(np.add(np.array([1,1,1,1,1]), r_4))

try:
    print(np.add(np.array([[1], [1]]), r_4))
except ValueError as e:
    print(f"error: {e}")
```

Broadcasting works by stretching smaller arrays to match the shape of largers ones.

## Rounding

```python
print(f"r_3 original: {r_3}")
print(f"trunc: {np.trunc(r_3)}")
print(f"fix: {np.fix(r_3)}")
print(f"floor: {np.floor(r_3)}")
print(f"ceil: {np.ceil(r_3)}")
print(f"round: {np.round(r_3, 2)}")
```

## Cumulative Operations

```python
print(f"r_4:\n{r_4}")
print(f"\nSum of all values: {np.sum(r_4)}")
print(f"\nSum along axis=0 (by column) : \n{np.sum(r_4, axis=0)}")
print(f"\nSum along axis=1 (by row) : \n{np.sum(r_4, axis=1)}")
print(f"\nCumulative sum:\n{np.cumsum(r_4)}")
print(f"\nCumulative sum along axis=0:\n{np.cumsum(r_4, axis=0)}")
```

## Grocery Problem

```python
# vegetables, fruits, snacks, beverages
bag1 = np.array([1, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])
print(f"Bag 1: {bag1}")
print(f"Bag 2: {bag2}")

#Total items per bag
totals = np.array([np.sum(bag1), np.sum(bag2)])
print(f"The first bag contains {totals[0]} items, and the second bag contains {totals[1]} items.")

#Total per category
category_totals = np.add(bag1, bag2)
print(f"There are {category_totals[0]} vegetables, {category_totals[1]} fruits, {category_totals[2]} boxes of snacks, and {category_totals[3]} beverages in the two grocey bags.")
```

## Teamwork

```python
# Monday through friday
# Ben missed Monday, Carla missed Wednesday - use 0 for missed days
ali = np.array([5, 4, 6, 4, 2])
ben = np.array([0, 5, 5, 3, 4])
carla = np.array([3, 6, 0, 3, 4])

# Compile into one array
all_knocks = np.array([ali, ben, carla])
print(all_knocks)

#Total per person and overall
per_person = np.sum(all_knocks, axis=1)
total = np.sum(all_knocks)
print(f"Ali knocked on {per_person[0]} doors, Ben knocked on {per_person[1]}, Carla knocked on {per_person[2]}.")
print(f"Total doors knocked: {total}")

# Rolling total of estimated votes (1 vote per 2 door knocks)
daily_totals = np.sum(all_knocks, axis=0)
rolling_votes = np.cumsum(daily_totals) * .5
print(f"Rolling estimated votes by day: {rolling_votes}")
```
