# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings - part codes

```python
#returns everything before colon
def get_supplier_code(part_code):
    return part_code[:part_code.index(':')]

#returns everything between the colon and dash
def get_product_num(part_code):
    return part_code[part_code.index(':') + 1 : part_code.index('-')]

#returns everything after the dash
def get_size(part_code):
    return part_code[part_code.index('-') + 1:]
```

```python
code1 = 'ACME:123-L'
code2 = 'DI:12345-M'
code3 = 'ACE:1-12'
```

```python
print(f"For {code1}, the supplier is {get_supplier_code(code1)}. Part number {get_product_num(code1)} is available in size {get_size(code1)}.")
```

```python
print(f"For {code2}, the supplier is {get_supplier_code(code2)}. Part number {get_product_num(code2)} is available in size {get_size(code2)}.")
```

```python
print(f"For {code3}, the supplier is {get_supplier_code(code3)}. Part number {get_product_num(code3)} is available in size {get_size(code3)}.")
```

```python
part_codes = ['ACME:123-L', 'DI:12345-M', 'ACE:1-12']

for code in part_codes:
    print(f"For {code}, the supplier is {get_supplier_code(code)}. Part number {get_product_num(code)} is available in size {get_size(code)}.")
```
