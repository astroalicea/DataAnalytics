# Exersice 5.A Numoy Fundamentals
This notebook covers the basics of NumPy arrays including creation, data types, dimensions, slicing, and transposing.

```python
import numpy as np
```

## Array Basics

```python
fruits = ['apple', 'banana', 'cherry', 'date']
```

```python
arr_1 = np.array(fruits)
print(arr_1)
```

```python
print(arr_1.dtype)
# U6 means Unicode string, max 6 characters long
```

```python
arr_2 = np.array([(10, 11, 12), (13, 14, 15)])
print(arr_2)
```

```python
print(arr_2.T)
# This flips rows and columns
```

```python
print(arr_2)
# arr_2 is unchanged. T creates a new view, it does not modify the original
```

```python
arr_3 = np.array([(1, 3, 5), (2, 4, 6)])
print(arr_3)
```

```python
arr_4 = np.array([arr_3[0, 2], arr_3[1, 1]])
print(arr_4)
#arr_3[0,2] = row 0, column 2 = 5
#arr_3[1,1] = row 1, column 1 = 4
#so arr_4 = [5,4]
```

```python
arr_5 = np.ones((2, 2, 2))
print(arr_5)
```
