## Lab 3. Tax functions
This notebook demonstrates calculating payroll tax withholdings using custon functions

```python
def get_soc_sec_tax(gross_pay):
    return gross_pay * .062
```

```python
def get_medicare_tax(gross_pay):
    return gross_pay * .0145
```

```python
def get_federal_tax(gross_pay, withholding_code):
    if withholding_code == 0:
        rate = .23
    elif withholding_code == 1:
        rate = .21
    elif withholding_code == 2:
        rate = .195
    elif withholding_code == 3:
        rate = .185
    elif withholding_code >= 4:
        #for every withholding over 4, subtract .5%
        rate = .18 - ((withholding_code - 4) * .005)
    return gross_pay * rate
```

```python
gross1 = 750
code1 = 0
print(f"Person 1 - Gross Pay: ${gross1:.2f}, Withholding Code: {code1}")
print(f" Social Security Tax: ${get_soc_sec_tax(gross1):.2f}")
print(f" Medicare Tax: ${get_medicare_tax(gross1):.2f}")
print(f" Federal Tax: ${get_federal_tax(gross1, code1):.2f}")
```
