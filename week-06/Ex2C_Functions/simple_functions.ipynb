# Exercise 2.C Creating and calling functions
## Lab 1. Working with defining custom functions

```python
def favorite_things(name, movie, music):
    return f"My name is {name}.\nMy favorite movie is {movie}.\nMy favorite musician is {music}."
```

```python
print(favorite_things('Luis', 'Once Upon a time in Hollywood', 'Hernan Cattaneo'))
```

```python
def why_im_here(purpose):
    return purpose
print(why_im_here('I joined the Year Up United Data Analyst program to build a career in tech.'))
```

```python
def favorite_place(place, why):
    return f"One of my favorite places is {place} beacuse {why}."
```

```python
print(favorite_place('Panama', 'the nature, city and food are amazing'))
```

```python
print(favorite_place('Spain', 'of the pretty cities and beaches'))
```
