# Exercise 2.B Using library functions
## Lab 2. Working with datetime functions

```python
import datetime
```

# Today

```python
today = datetime.datetime.now()
print(today.strftime("Today is %A, %B %d, %Y"))
```

# My birthday

```python
birthday = datetime.date(1997, 2, 13)
```

```python
print(birthday.strftime("My birthday is %m/%d/%Y"))
```

# Next 90 Days

```python
dinner_time = datetime.time(7, 30)
print(dinner_time.strftime("Lets meet for dinner at %I:%M %p"))
```
